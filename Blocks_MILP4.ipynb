```python
import gurobipy as gp
import os
import shutil
import json, os, collections

class MILPFindReboundPath:
    """
    MILPFindReboundPath
    ä½¿ç”¨MILPå¯»æ‰¾Reboundå‹å¥½çš„æˆªæ–­å·®åˆ†è·¯çº¿
    """

    def __init__(self, ciphername, rounds, begin_round, end_round) -> None:
        self.__model = gp.Model()
        self.__model.setParam(gp.GRB.Param.OutputFlag, False)

        self.__ciphername = ciphername
        self.__rounds = rounds

        self.__num_blocks = 12
        self.__block_size = 16
        # self.__state_size = 16
        self.__state_size = self.__num_blocks * self.__block_size ##192

        self.__key_size = 16 * 12
        self.__state_vars = []
        self.__auxi_vars = []
        self.__aes_type = 128 * 12 #########æ³¨æ„è¿™é‡Œæ”¹ä¹˜12æ”¾å®½äº†çº¦æŸ

        self.__InboundCeil = 16 * 12 * 3
        self.__DisCeil = 8 * 16 * 12 * 5
        self.__LastCeil = 16 * 3 #####################ä¸ç”¨ä¹˜8
        self.__ColiCeil = 8 * 16 * 12 * 4
        self.__XorMatchLimit = 16 * 4

        self.__inbound_begin_round = begin_round
        self.__inbound_end_round = end_round
        # block activation matrix: é»˜è®¤å…¨éƒ¨æ¿€æ´»ï¼ˆrounds x num_blocksï¼‰
        # å¯ä»¥é€šè¿‡ set_block_activation æ–¹æ³•è®¾ç½®ä¸ºåªéƒ¨åˆ†å—æ¿€æ´»
        self.__block_active = [[True for _ in range(self.__num_blocks)] for _ in range(self.__rounds)]

        self.__milp_lpfile_folder = ""
        self.__milp_lpfile_name = ""

        self.__xor_match_vars = []


        # === åˆå§‹åŒ– fixed_mask ===
        # self.__fixed_mask = {}
        # r = 0
        # print(self.__fixed_mask)

        # æ•´å—ç½® 0
        # for b in [0, 1, 2, 3, 6, 7]:
        # for b in [3, 6, 7]:
        #     for i in range(self.__block_size):
        #         self.__fixed_mask[(r, b, i)] = 0

        # å‰©ä½™å—ï¼šindex=0 -> 1ï¼Œå…¶ä½™ -> 0
        # for b in range(self.__num_blocks):
        #     if b not in [0, 1, 2, 3, 6, 7]:
        #         for i in range(self.__block_size):
        #             if i == 0:
        #                 self.__fixed_mask[(r, b, i)] = 1
        #             else:
        #                 self.__fixed_mask[(r, b, i)] = 0


    def generate_variables(self) -> None:
        model = self.__model
        rounds = self.__rounds
        state_size = self.__state_size
        num_blocks = self.__num_blocks
        block_size = self.__block_size
        num_cols = 4

        # æ¸…ç†æ—§å˜é‡ï¼ˆé˜²æ­¢é‡å¤è°ƒç”¨æ—¶è¿½åŠ ï¼‰
        self.__state_vars = []
        self.__auxi_vars = []

        # for r in range(rounds+1): ###########è®©å¤šå¾ªç¯ä¸€è½®
        #     round_vars = []
        #     for b in range(num_blocks):
        #         block_vars = []
        #         for i in range(block_size):
        #             block_vars.append(model.addVar(vtype=gp.GRB.BINARY, name=f"x_{r}_{b}_{i}"))
        #         round_vars.append(block_vars)
        #     self.__state_vars.append(round_vars)

        fixed_mask = [[[-1 for _ in range(block_size)] for _ in range(num_blocks)] for _ in range(rounds+1)]
        # for b in range(4):
        #     for i in range(16):
        #         fixed_mask[0][b][i] = 0
        ###########################è®©å‰å››ä¸ªå—æ˜¯0

        for r in range(rounds):
            round_vars = []
            for b in range(num_blocks):
                block_vars = []
                for i in range(block_size):
                    # if (r, b, i) in fixed_mask:
                    #     # å›ºå®šå¸¸é‡ï¼šç›´æ¥ç”¨ä¸€ä¸ªâ€œè™šæ‹Ÿå˜é‡â€ä»£æ›¿
                    #     const_val = fixed_mask[(r, b, i)]
                    #     print(const_val)
                    #     v = model.addVar(vtype=gp.GRB.BINARY, name=f"x_{r}_{b}_{i}")
                    #     model.addConstr(v == const_val)
                    # else:
                    v = model.addVar(vtype=gp.GRB.BINARY, name=f"x_{r}_{b}_{i}")
                    block_vars.append(v)
                    # if fixed_mask is not None:
                    if fixed_mask[r][b][i] == 0:
                        model.addConstr(v == 0, name=f"fix0_{r}_{b}_{i}_0")
                    elif fixed_mask[r][b][i] == 1:
                        model.addConstr(v == 1, name=f"fix1_{r}_{b}_{i}_1")

                round_vars.append(block_vars)
            self.__state_vars.append(round_vars)

        for r in range(rounds - 1): #######è¾…åŠ©å˜é‡ä¹Ÿå¤šå¾ªç¯ä¸€è½®
            round_auxi = []
            for b in range(num_blocks):
                block_auxi = []
                for c in range(num_cols):
                    block_auxi.append(model.addVar(vtype=gp.GRB.BINARY, name=f"d_{r}_{b}_{c}"))
                round_auxi.append(block_auxi)
            self.__auxi_vars.append(round_auxi)

        model.update()
        print(self.__state_vars)

    def mds_cocnstraints(self, var_in, var_out, auxi_var) -> None:
        model = self.__model
        model.addConstr(sum(var_in) + sum(var_out) - 5 * auxi_var >= 0)
        model.addConstr(sum(var_in) - auxi_var >= 0)
        model.addConstr(sum(var_out) - auxi_var >= 0)
        for i in range(4):
            model.addConstr(auxi_var - var_in[i] >= 0)
            model.addConstr(auxi_var - var_out[i] >= 0)
        model.update()

    def aes_round(self, index_in:int, index_out: int, r: int) -> None:
        model = self.__model
        rounds = self.__rounds
        state_size = self.__state_size
        state_vars = self.__state_vars
        block_in = self.__state_vars[r][index_in]
        block_out = self.__state_vars[r + 1][index_out]
        auxi_block = self.__auxi_vars[r][index_out]

        index = [[0, 5, 10, 15], [4, 9, 14, 3], [8, 13, 2, 7], [12, 1, 6, 11]]
        # for r in range(rounds - 1):
        for i in range(4):
            var_in = []
            var_out = []
            for j in range(4):
                var_in.append(block_in[index[i][j]])
            for j in range(4):
                var_out.append(block_out[4 * i + j])
            self.mds_cocnstraints(var_in, var_out, self.__auxi_vars[r][index_out][i])

        # for i in range(4):
        #     var_in = []
        #     var_out = []
        #     for j in range(4):
        #         var_in.append(block_in[4*i + j])
        #     for j in range(4):
        #         var_out.append(block_out[4*i + j])
        #     self.mds_cocnstraints(var_in, var_out, self.__auxi_vars[r][index_out][i])


    def aes_temp(self, block_in, block_out, index_out, r) -> None:
        model = self.__model
        rounds = self.__rounds
        state_size = self.__state_size
        state_vars = self.__state_vars

        index = [[0, 5, 10, 15], [4, 9, 14, 3], [8, 13, 2, 7], [12, 1, 6, 11]]
        # for r in range(rounds - 1):
        for i in range(4):
            var_in = []
            var_out = []
            for j in range(4):
                var_in.append(block_in[index[i][j]])
            for j in range(4):
                var_out.append(block_out[4 * i + j])
            self.mds_cocnstraints(var_in, var_out, self.__auxi_vars[r][index_out][i])

        # for i in range(4):
        #     var_in = []
        #     var_out = []
        #     for j in range(4):
        #         var_in.append(block_in[4*i + j])
        #     for j in range(4):
        #         var_out.append(block_out[4*i + j])
        #     self.mds_cocnstraints(var_in, var_out, self.__auxi_vars[r][index_out][i])


    def xor_block_constraints(self, r, index_in1, index_in2, index_out) -> None:
        """
        ä¸ºä¸¤ä¸ªå—çš„ XOR æ„å»º MILP çº¦æŸ
        block_in1, block_in2, block_out: åˆ—è¡¨ é•¿åº¦ä¸º16 æ¯ä¸ªå…ƒç´ æ˜¯GurobiäºŒè¿›åˆ¶å˜é‡
        """
        # model = self.__model
        # for i in range(16):
        #     x = block_in1[i]
        #     y = block_in2[i]
        #     z = block_out[i]

        #     # XOR æ´»è·ƒæ€§çº¦æŸ
        #     model.addConstr(z >= x)
        #     model.addConstr(z >= y)
        #     model.addConstr(z <= x + y) ###############ä¸€èˆ¬ä¼¼ä¹ä¹Ÿä¸è€ƒè™‘1+1=0çš„æƒ…å†µï¼Œé»˜è®¤1+1æ˜¯1
        
        # model.update()
        model = self.__model
        state_vars = self.__state_vars


        block_in1 = state_vars[r][index_in1]
        block_in2 = state_vars[r][index_in2]
        block_out = state_vars[r + 1][index_out]

        for i in range(16):
            x = block_in1[i]
            y = block_in2[i]
            z = block_out[i]

            # XOR æ´»è·ƒæ€§çº¦æŸæƒ³ 
            model.addConstr(z >= x - y)
            model.addConstr(z >= y - x)
            model.addConstr(z <= x + y)  # é»˜è®¤1+1=1
            ##########################################è¿™é‡Œå¼‚æˆ–çš„é€»è¾‘æ˜¯1+1ä¸€å®šå‡º1ï¼Œå¯ä»¥æŠŠè¿™éƒ¨åˆ†å»ºæ¨¡ä¸€ä¸‹

            # è®°å½• "1+1=0" çš„æƒ…å†µ
            m = model.addVar(vtype=gp.GRB.BINARY, name=f"xor_match_{r}_{index_in1}_{index_in2}_{index_out}_{i}")
            model.addConstr(m >= x - z)
            model.addConstr(m >= y - z)
            model.addConstr(m <= 1 - z)
            model.addConstr(m <= x + y - z)
            # ä¿å­˜åˆ°å…¨å±€åˆ—è¡¨
            self.__xor_match_vars.append(m)
        model.update()

    def xor_block_constraints_tmp(self, r, block_in1, block_in2, block_out) -> None:
        """
        ä¸ºä¸¤ä¸ªå—çš„ XOR æ„å»º MILP çº¦æŸ
        block_in1, block_in2, block_out: åˆ—è¡¨ é•¿åº¦ä¸º16 æ¯ä¸ªå…ƒç´ æ˜¯GurobiäºŒè¿›åˆ¶å˜é‡
        """
        model = self.__model
        for i in range(16):
            x = block_in1[i]
            y = block_in2[i]
            z = block_out[i]

            # XOR æ´»è·ƒæ€§çº¦æŸ
            model.addConstr(z >= x)
            model.addConstr(z >= y)
            model.addConstr(z <= x + y) ###############ä¸€èˆ¬ä¼¼ä¹ä¹Ÿä¸è€ƒè™‘1+1=0çš„æƒ…å†µï¼Œé»˜è®¤1+1æ˜¯1

            m = model.addVar(vtype=gp.GRB.BINARY)
            model.addConstr(m >= x - z)
            model.addConstr(m >= y - z)
            model.addConstr(m <= 1 - z)
            model.addConstr(m <= x + y - z)
            # ä¿å­˜åˆ°å…¨å±€åˆ—è¡¨
            self.__xor_match_vars.append(m)
            
        model.update()

    def aes_then_xor(self, index_in: int, index_in2: int, index_out: int, r: int) -> None:
        """
        block_in  : å…ˆè¿‡ AES çš„å—
        block_in2 : ä¸ AES è¾“å‡º XOR çš„å—
        block_out : è¾“å‡ºå—
        r         : å½“å‰è½®å·
        """
        # 1. AES è¾“å‡ºåˆ°ä¸´æ—¶å—
        # block_tmp = [self.__model.addVar(vtype=gp.GRB.BINARY, name=f"tmp_r{r}_{i}") for i in range(16)]
        # self.aes_round(block_in, block_tmp, r)  # AESçº¦æŸ

        # # 2. XOR ä¸´æ—¶å—ä¸ block_in2ï¼Œå¾—åˆ°æœ€ç»ˆè¾“å‡º
        # self.xor_block_constraints(block_tmp, block_in2, block_out)
        # 1. AES è¾“å‡ºåˆ°ä¸´æ—¶å—
        model = self.__model
        rounds = self.__rounds
        state_size = self.__state_size
        state_vars = self.__state_vars

        # block_tmp = [self.__model.addVar(vtype=gp.GRB.BINARY, name=f"tmp_r{r}_{i}") for i in range(16)]
        block_tmp = self._fix_tmp_naming_and_create_tmp(r, index_out)

        block_in = self.__state_vars[r][index_in]
        block_out = self.__state_vars[r + 1][index_out]
        block_in2 = self.__state_vars[r][index_in2]

        self.aes_temp(block_in, block_tmp, index_out, r)
        self.xor_block_constraints_tmp(r, block_tmp, block_in2, block_out)

    ########################################################################################################################### 
    # ----------------- Debug helpers & small fixes -----------------

    def _get_var_value(self, v):
        # safe getter: æ”¯æŒå•è§£ (.X) å’Œè§£æ±  (.xn)
        if v is None:
            return None
        return getattr(v, "X", getattr(v, "xn", None))

    def _fix_tmp_naming_and_create_tmp(self, r, index_out):
        """ç”¨äºåœ¨ aes_then_xor ä¸­åˆ›å»ºå”¯ä¸€å‘½åçš„ tmp å˜é‡ï¼ˆé¿å…é‡å¤åå­—å¯¼è‡´æ··æ·†ï¼‰"""
        model = self.__model
        tmp = [model.addVar(vtype=gp.GRB.BINARY, name=f"tmp_r{r}_out{index_out}_{i}") for i in range(16)]
        return tmp

    def debug_aes_round_print(self, r, index_in, index_out):
        """
        è§£åè°ƒç”¨ï¼šæ‰“å° aes_round åœ¨ round r (ä» r -> r+1) å¯¹åº”çš„ä¸€ç»„ 4 åˆ—çš„
        - æ¯åˆ—çš„è¾“å…¥ç´¢å¼• (åœ¨ block_in å†…)
        - å¯¹åº” input bit çš„è§£å€¼
        - å¯¹åº” output bit çš„è§£å€¼
        - å¯¹åº” auxi_var çš„è§£å€¼
        è¿™æ ·èƒ½ç«‹åˆ»çœ‹å‡ºï¼šSR æ˜ å°„æ˜¯å¦æ­£ç¡®ã€MC æ˜¯å¦â€œæŠŠâ€æ´»è·ƒæ‰©æ•£åˆ°å…¶å®ƒåˆ—
        """
        index = [[0,5,10,15],[4,9,14,3],[8,13,2,7],[12,1,6,11]]
        getv = self._get_var_value

        print(f"\n>>> Debug AES round r={r}: block_in={index_in} -> block_out={index_out} <<<")
        # print variable names for clarity
        print(" block_in varnames:")
        print([v.VarName for v in self.__state_vars[r][index_in]])
        print(" block_out varnames:")
        print([v.VarName for v in self.__state_vars[r+1][index_out]])
        print(" auxi varnames:")
        print([v.VarName for v in self.__auxi_vars[r][index_out]])

        for col in range(4):
            in_indices = index[col]
            in_vars = [self.__state_vars[r][index_in][k] for k in in_indices]
            in_vals = [int(getv(v) or 0) for v in in_vars]

            out_indices = [4*col + j for j in range(4)]
            out_vars = [self.__state_vars[r+1][index_out][k] for k in out_indices]
            out_vals = [int(getv(v) or 0) for v in out_vars]

            aux = int(getv(self.__auxi_vars[r][index_out][col]) or 0)

            print(f" col {col}: in_idx={in_indices}, in_vals={in_vals}, sum_in={sum(in_vals)} | "
                f"out_idx={out_indices}, out_vals={out_vals}, sum_out={sum(out_vals)} | aux={aux}")

    def debug_print_all_rounds(self):
        """ä¸€é”®æ‰“å°æ•´ä¸ªè§£ï¼ˆæ¯è½®æ¯ block çš„ 16 ä¸ªæ¯”ç‰¹ï¼‰â€”â€”è§£æ¯”è¾ƒå°æ—¶å¯ç”¨"""
        getv = self._get_var_value
        for r in range(self.__rounds):
            print(f"\n=== Round {r} state blocks ===")
            for b in range(self.__num_blocks):
                vals = [int(getv(self.__state_vars[r][b][i]) or 0) for i in range(16)]
                print(f" r{r} b{b}: {vals}")

    def check_mds_consistency(self):
        """
        æ£€æŸ¥æ¯ä¸ª aes_round çš„æ¯åˆ—æ˜¯å¦æ»¡è¶³ 'å¦‚æœè¾“å…¥æ¿€æ´»åˆ™è¾“å‡ºè‡³å°‘æœ‰ä¸€ä¸ªæ¿€æ´»'ï¼ˆä»¥åŠåå‘ï¼‰
        æŠŠæ‰€æœ‰è¿åé¢„æœŸçš„åˆ—æ‰“å°å‡ºæ¥ï¼ˆæœ‰åŠ©äºå®šä½çº¦æŸæ˜¯å¦è¢«é”™è¯¯å†™æ­»æˆ–ç´¢å¼•é”™ä½ï¼‰
        """
        index = [[0,5,10,15],[4,9,14,3],[8,13,2,7],[12,1,6,11]]
        getv = self._get_var_value
        bad = []
        for r in range(self.__rounds - 1):
            for b_out in range(self.__num_blocks):
                # this assumes auxi_vars exist for this (r, b_out)
                for col in range(4):
                    in_idx = index[col]
                    # è¯´æ˜ï¼šè°æ˜¯å¯¹åº”çš„ block_inï¼Ÿæˆ‘ä»¬åªèƒ½çŒœæ˜¯ï¼šblock_in è¢« aes_round ç”¨åˆ°çš„ä½ç½®å–å†³äºä½ Permutationè°ƒç”¨
                    # è¿™é‡Œå…ˆ brute-force æ£€æŸ¥ï¼šå¯¹äºå½“å‰ rï¼Œæ‰«ææ‰€æœ‰å¯èƒ½ block_inï¼Œçœ‹å“ªä¸ªæ»¡è¶³ auxi reference
                    sum_in = None
                    sum_out = sum(int(getv(self.__state_vars[r+1][b_out][4*col + j]) or 0) for j in range(4))
                    aux_val = int(getv(self.__auxi_vars[r][b_out][col]) or 0)
                    # è‹¥ aux=1ï¼Œä½† sum_out==0 -> å¯ç–‘
                    if aux_val == 1 and sum_out == 0:
                        bad.append((r, b_out, col, "aux=1 but sum_out=0"))
                    # è‹¥ sum_out>0 but aux==0 -> å¯ç–‘
                    if sum_out > 0 and aux_val == 0:
                        bad.append((r, b_out, col, "sum_out>0 but aux=0"))
        if not bad:
            print("MDS consistency quick-check: all OK (no aux/output mismatch found).")
        else:
            print("MDS consistency issues found:")
            for item in bad:
                print(" ", item)

    # ----------------- ä¿®æ”¹ aes_then_xor å†… tmp åˆ›å»ºï¼ˆç¤ºä¾‹ï¼‰ -----------------
    # åœ¨ä½ çš„ aes_then_xor ä¸­ï¼ŒæŠŠåŸæ¥
    #    block_tmp = [self.__model.addVar(..., name=f"tmp_r{r}_{i}") for i in range(16)]
    # æ”¹ä¸ºï¼š
    #    block_tmp = self._fix_tmp_naming_and_create_tmp(r, index_out)
    # è¿™æ · tmp åç§°åŒ…å« index_outï¼Œé¿å…åŒä¸€ roundã€ä¸åŒ block è°ƒç”¨æ—¶é‡åï¼Œåˆ©äº debugã€‚
#########################################################################################################################

    def Permutation_round(self, r) -> None:
        model = self.__model
        state_size = self.__state_size
        state_vars = self.__state_vars
        # state_in = state_vars[r]
        # state_out = state_vars[r + 1]

        self.aes_then_xor(2, 8, 0, r)
        self.xor_block_constraints(r, 0, 9, 1)
        self.aes_then_xor(3, 10, 2, r)
        self.aes_then_xor(1, 4, 3, r)
        self.xor_block_constraints(r, 6, 11, 4)
        self.xor_block_constraints(r, 7, 8, 5)
        self.xor_block_constraints(r, 2, 3, 6)
        self.xor_block_constraints(r, 1, 5, 7)
        self.aes_then_xor(6, 4, 8, r)
        self.aes_round(0, 9, r)
        self.aes_then_xor(7, 9, 10, r)
        self.xor_block_constraints(r, 5, 10, 11)

    # def Permutation_Inv() -> None:
        
    def Permutation(self) -> None:
        """
        å®šä¹‰æ¯è½®çš„å—çº§ç½®æ¢
        """
        for r in range(self.__rounds - 1):
            self.Permutation_round(r)

    # def set_object_function(self) -> None:
    #     equation = []
    #     for i in range(self.__rounds):
    #         equation.append(sum(self.__state_vars[i]))
    #     self.__model.setObjective(sum(equation), gp.GRB.MINIMIZE)
    #     self.__model.update()

    def set_object_function(self) -> None:
        """
        ç›®æ ‡å‡½æ•°: æœ€å°åŒ–æ‰€æœ‰å—çš„æ¿€æ´»æ•°é‡
        """
        model = self.__model
        obj_terms = []
        for r in range(self.__rounds):
            for b in range(self.__num_blocks):
                obj_terms.extend(self.__state_vars[r][b])  # æ¯ä¸ªå—16ä¸ªæ¯”ç‰¹
        model.setObjective(gp.quicksum(obj_terms), gp.GRB.MINIMIZE)
        model.update()

    # def init_constraints(self) -> None:
    #     self.__model.addConstr(sum(self.__state_vars[0]) >= 1)
    #     self.__model.update()

    def init_constraints(self) -> None:
        """
        åˆå§‹è½®è‡³å°‘æœ‰ä¸€ä¸ªæ¯”ç‰¹è¢«æ¿€æ´»
        """
        model = self.__model
        first_round_terms = []
        for b in range(self.__num_blocks):
            first_round_terms.extend(self.__state_vars[0][b])
        model.addConstr(gp.quicksum(first_round_terms) >= 1)
        model.update()


    # def inbound_constraints(self) -> None:
    #     equation = []
    #     begin_round, end_round = self.__inbound_begin_round, self.__inbound_end_round
    #     for r in range(begin_round, end_round + 1):
    #         equation.append(sum(self.__state_vars[r]))
    #     self.__model.addConstr(sum(equation) <= self.__state_size + self.__key_size)
    #     self.__model.update()

    
    def inbound_constraints(self) -> None:
        """
        é™åˆ¶ inbound é˜¶æ®µæ€»æ´»è·ƒæ¯”ç‰¹æ•°
        """
        model = self.__model
        InboundCeil = self.__InboundCeil

        begin_round, end_round = self.__inbound_begin_round, self.__inbound_end_round
        inbound_terms = []
        for r in range(begin_round, end_round + 1):
            for b in range(self.__num_blocks):
                inbound_terms.extend(self.__state_vars[r][b])
        # model.addConstr(gp.quicksum(inbound_terms) <= self.__state_size + self.__key_size) #######è¿™é‡Œçš„state_sizeæ˜¯192
        model.addConstr(gp.quicksum(inbound_terms) <= InboundCeil)
        model.update()




    def outbound_constraints(self) -> None:
        """
        é™åˆ¶ outbound é˜¶æ®µè¾…åŠ©å˜é‡å’Œå—æ´»è·ƒæ€§
        """
        
        model = self.__model
        if hasattr(self, "_outbound_constraints_list"):
            for c in self._outbound_constraints_list:
                model.remove(c)
            model.update()
        begin_round, end_round = self.__inbound_begin_round, self.__inbound_end_round


        # # å‰å‘é˜¶æ®µ (0 åˆ° begin_round-2)
        # forward_auxi_sum = 0
        # for r in range(begin_round - 1):
        #     for b in range(12):#############è¿™é‡Œä»4æ”¹æˆäº†12ï¼Œä¸‹é¢ä¹Ÿæ˜¯
        #         # æ¯åˆ—4æ¯”ç‰¹
        #         col_sum = gp.quicksum(self.__state_vars[r][b][i] for i in range(16))
        #         forward_auxi_sum += self.__auxi_vars[r][b][0] * (4 - col_sum)  # MDSçº¦æŸå¯¹åº”çš„è¾…åŠ©å˜é‡
        # #####################################Outboundçº¦æŸçš„æ˜¯ç™½æ ¼å­éæ´»è·ƒå­—èŠ‚
        # # åå‘é˜¶æ®µ (end_round+2 åˆ° rounds-2)
        # back_auxi_sum = 0
        # back_auxi_vars = self.__auxi_vars[end_round + 1 : self.__rounds - 1]
        # for r, auxi_block in enumerate(back_auxi_vars):
        #     for b in range(12):
        #         col_sum = gp.quicksum(self.__state_vars[end_round + 2 + r][b][i] for i in range(16))
        #         back_auxi_sum += auxi_block[b][0] * (4 - col_sum)

        # å‰å‘é˜¶æ®µ (0 åˆ° begin_round-2)
        forward_auxi_sum = 0
        for r in range(begin_round - 1):
            for b in range(12):  # éå†æ¯ä¸ª block
                for c in range(4):  # éå† block çš„æ¯ä¸€åˆ—
                    col_bits = self.__state_vars[r][b][4*c : 4*c + 4]  # å–å‡ºè¿™ä¸€åˆ—çš„4æ¯”ç‰¹
                    col_sum = gp.quicksum(col_bits)
                    forward_auxi_sum += self.__auxi_vars[r][b][c] * (4 - col_sum)

        # åå‘é˜¶æ®µ (end_round+2 åˆ° rounds-2)
        back_auxi_sum = 0
        back_auxi_vars = self.__auxi_vars[end_round + 1 : self.__rounds - 1]
        for r, auxi_block in enumerate(back_auxi_vars):
            for b in range(12):  # éå†æ¯ä¸ª block
                for c in range(4):  # éå† block çš„æ¯ä¸€åˆ—
                    col_bits = self.__state_vars[end_round + 2 + r][b][4*c : 4*c + 4]
                    col_sum = gp.quicksum(col_bits)
                    back_auxi_sum += auxi_block[b][c] * (4 - col_sum)


        # è®¾ç½®ç›®æ ‡å‡½æ•°å’Œçº¦æŸ
        # aes_type = 128 * 12 * 4#########æ³¨æ„è¿™é‡Œæ”¹ä¹˜12æ”¾å®½äº†çº¦æŸï¼Œæœ€åä¹˜ä¸€ä¸ªæ”¾å¤§ç³»æ•°ï¼Œä¸ºäº†æœ‰è§£
        DisCeil = self.__DisCeil
        LastCeil = self.__LastCeil #############ä¸ç”¨ä¹˜8
        ColiCeil = self.__ColiCeil

        first_round_sum = gp.quicksum(self.__state_vars[0][b][i] for b in range(self.__num_blocks) for i in range(16))
        last_round_sum = gp.quicksum(self.__state_vars[self.__rounds - 1][b][i] for b in range(self.__num_blocks) for i in range(16))
        ###############åŸæ–‡é‡Œä¸é™åˆ¶è¾“å‡ºå±‚çš„æ´»è·ƒæ˜¯å› ä¸ºæ‹›ç¢°æ’ä¸‹é¢æœ‰SRåå¯¹é½çš„ä»£ç 
        #########################åˆå§‹è½®ä¹Ÿæ˜¯å¸Œæœ›æ´»è·ƒå­—èŠ‚æœ€å°åŒ–çš„

        
        ##############################################æŒ‰åŒºåˆ†å™¨æ€è·¯æ‰¾ï¼Œé™åˆ¶æœ€åä¸€è½®çš„æ´»è·ƒSç›’
        model.setObjective(8 * (forward_auxi_sum + back_auxi_sum + first_round_sum + last_round_sum), gp.GRB.MINIMIZE) ###æ•´æ¡è·¯å¾„çš„æ¿€æ´»å­—èŠ‚æœ€å°åŒ–
        model.addConstr(8 * (forward_auxi_sum + back_auxi_sum + first_round_sum + last_round_sum) <= int(DisCeil)) ######æ¿€æ´»å­—èŠ‚è®¾ç½®äº†ä¸Šé™###æ³¨æ„
        model.addConstr(last_round_sum <= LastCeil)  ########åˆå§‹è½®è‡³å°‘æœ‰ä¸€ä¸ªæ¿€æ´»å­—èŠ‚


        ############################################æŒ‰ç¢°æ’æ€è·¯æ‰¾ï¼Œå¼ºåˆ¶ç¬¬ä¸€è½®å’Œæœ€åä¸€è½®è¿‡SRåå¯¹é½
        # model.setObjective(8 * (forward_auxi_sum + back_auxi_sum + first_round_sum), gp.GRB.MINIMIZE) ###æ•´æ¡è·¯å¾„çš„æ¿€æ´»å­—èŠ‚æœ€å°åŒ–
        # model.addConstr(8 * (forward_auxi_sum + back_auxi_sum + first_round_sum) <= ColiCeil) ######æ¿€æ´»å­—èŠ‚è®¾ç½®äº†ä¸Šé™###æ³¨æ„
        # # ShiftRowçº¦æŸ: è¾“å…¥è½® == è¾“å‡ºè½®#####################è¿™é‡Œå¥‡æ€ªï¼Œå¯¹äºæ„é€ ç¢°æ’æœ‰ç”¨ï¼Œå¼ºåˆ¶å¯¹é½
        # begin_state = self.__state_vars[0]
        # end_state = self.__state_vars[self.__rounds - 1]
        # index_sr = [0, 5, 10, 15, 4, 9, 14, 3, 8, 13, 2, 7, 12, 1, 6, 11]
        # for b in range(self.__num_blocks):
        #     for i in range(16):
        #         model.addConstr(begin_state[b][i] == end_state[b][index_sr[i]])

        model.update()

    # def construct_active_sbox_model(self) -> None:
    #     self.generate_variables()
    #     self.aes_round()
    #     self.set_object_function()
    #     self.init_constraints()

    #     folder_path = "AES_active_sbox_milp_lp"
    #     filename = f"AES{self.__aes_type}_round_{self.__rounds}.lp"
    #     if not os.path.exists(folder_path):
    #         os.makedirs(folder_path)
    #     self.__model.write(f"{folder_path}/{filename}")

    def construct_active_sbox_model(self) -> None:
        """
        æ„é€ ä¸€ä¸ªæ™®é€šçš„ active-sbox MILP å¯¹æ‰€æœ‰è½®ã€å¯¹æ‰€æœ‰å—æ‰§è¡Œ AES å— -> åŒå—ï¼‰
        è¿™æ ·å¯ä»¥å¾—åˆ°æœ€å°æ´»è·ƒ S-box çš„åŸºçº¿æ¨¡å‹ã€‚
        """
         # ========== 1) é‡ç½®æ¨¡å‹ï¼ˆé¿å…é‡å¤å˜é‡/çº¦æŸï¼‰ ==========
        self.__model = gp.Model()
        self.__model.setParam(gp.GRB.Param.OutputFlag, False)
        # ç”Ÿæˆå˜é‡
        self.generate_variables()

        # å¯¹æ¯ä¸ª r ä» 0 åˆ° rounds-2ï¼Œ å¯¹æ¯ä¸ªå— b åš aes_round(b -> b)
        # for r in range(self.__rounds - 1):
        self.Permutation()

        # ç›®æ ‡ + åŸºæœ¬çº¦æŸ
        self.set_object_function()
        self.init_constraints() ###########è¿™éƒ¨åˆ†ä»£ç æ²¡æœ‰å†™å…¥æ–‡ä»¶

    # def construct_rebound_friend_model(self) -> None:
    #     self.generate_variables()
    #     self.aes_round()
    #     self.inbound_constraints()
    #     self.outbound_constraints()
    #     self.init_constraints()

    #     self.__milp_lpfile_folder = f"{self.__ciphername}_rebound_milp_lp"
    #     self.__milp_lpfile_name = f"AES{self.__aes_type}_round_{self.__rounds}.lp"
    #     if not os.path.exists(self.__milp_lpfile_folder):
    #         os.makedirs(self.__milp_lpfile_folder)
    #     self.__model.write(f"{self.__milp_lpfile_folder}/{self.__milp_lpfile_name}")

    def construct_rebound_friend_model(self) -> None:
        """
        æ„é€  Rebound-friendly çš„å®Œæ•´æ¨¡å‹ï¼š
        - å…ˆç”Ÿæˆå˜é‡
        - ç„¶åæ¯è½®æ‰§è¡Œ Permutation(r)ï¼ˆä½ å®šä¹‰çš„å—çº§ AES/XOR è°ƒç”¨åºåˆ—ï¼‰
        - å†åŠ å…¥ inbound/outbound/init ç­‰çº¦æŸ
        - å†™å…¥ lp
        """
        # ç”Ÿæˆå˜é‡ï¼ˆæ¸…ç†æ—§çš„ï¼‰
        self.generate_variables()#######################################################################################################################################
        # å¯¹æ¯ä¸€è½® r æ„å»ºè¯¥è½®çš„ç½®æ¢çº¦æŸï¼ˆPermutation æ˜¯ä½ ä¹‹å‰å®šä¹‰å¥½çš„å—çº§ç»„åˆï¼‰########åŸå‡½æ•°ä¹Ÿæ˜¯åªèµ°äº†ä¸€è½®
        # for r in range(self.__rounds - 1):
        self.Permutation()
                # åŠ å…¥ inbound/outbound/init ç­‰çº¦æŸï¼ˆåŸºäºæ–°çš„ block ç»“æ„ï¼‰
        self.inbound_constraints()
        self.outbound_constraints()
        self.init_constraints()
        
        # ä¿å­˜ lp æ–‡ä»¶
        self.__milp_lpfile_folder = f"{self.__ciphername}_rebound_milp_lp"
        self.__milp_lpfile_name = f"AES{self.__aes_type}_round_{self.__rounds}.lp"
        if not os.path.exists(self.__milp_lpfile_folder):
            os.makedirs(self.__milp_lpfile_folder)
        self.__model.write(f"{self.__milp_lpfile_folder}/{self.__milp_lpfile_name}")

    # def remove_foler(self):
    #     shutil.rmtree(self.folder_path)

    def remove_folder(self):
        """åˆ é™¤ä¸Šæ¬¡å†™å…¥çš„æ¨¡å‹æ–‡ä»¶å¤¹ï¼ˆå®‰å…¨æ£€æŸ¥å­˜åœ¨æ€§ï¼‰"""
        if hasattr(self, "_MILPFindReboundPath__milp_lpfile_folder"):
            folder = self.__milp_lpfile_folder
        else:
            folder = self.__milp_lpfile_folder
        if folder and os.path.exists(folder):
            shutil.rmtree(folder)

    
    def stop_at_gap(self, model, where) -> None:
        if where == gp.GRB.Callback.MIP:
            objbst = model.cbGet(gp.GRB.Callback.MIP_OBJBST)   # å½“å‰æœ€å¥½å¯è¡Œè§£
            objbnd = model.cbGet(gp.GRB.Callback.MIP_OBJBND)   # å½“å‰ä¸‹ç•Œ
            if objbst < 1e20 and abs(objbst) > 1e-9:           # ç¡®ä¿æœ‰è§£
                gap = abs(objbst - objbnd) / (abs(objbst) + 1e-10)
                if gap < 0.20:  # ç¬¬ä¸€æ¬¡å°äº 20%
                    print(f"Gap < 30% æ—¶åœæ­¢: æœ€ä¼˜è§£ = {objbst}, ä¸‹ç•Œ = {objbnd}, gap = {gap:.2%}")
                    model.terminate()  # ç»ˆæ­¢æ±‚è§£

    

    def my_callback(self, model, where):
        if where == gp.GRB.Callback.MIPSOL:
            obj = model.cbGet(gp.GRB.Callback.MIPSOL_OBJ)
            best = getattr(model, "_bestobj", None)

            # åªæœ‰æ›´ä¼˜è§£æ‰ä¿å­˜
            if best is None or obj < best - 1e-6:
                model._bestobj = obj

                sol_count = model.cbGet(gp.GRB.Callback.MIPSOL_SOLCNT)
                vars = model.getVars()
                sol = model.cbGetSolution(vars)

                solution_dict = {v.VarName: sol[i] for i, v in enumerate(vars)}

                # ä¿å­˜æ–‡ä»¶
                filename = f"checkpoint_{sol_count}.json"
                with open(filename, "w") as f:
                    json.dump(solution_dict, f, indent=2)

                # ç»´æŠ¤æ–‡ä»¶é˜Ÿåˆ—ï¼ˆæœ€å¤š5ä¸ªï¼‰
                if not hasattr(model, "_saved_files"):
                    model._saved_files = collections.deque(maxlen=5)

                model._saved_files.append(filename)

                # å¦‚æœè¶…è¿‡ 5 ä¸ªï¼Œdeque ä¼šè‡ªåŠ¨ä¸¢æ‰æœ€è€çš„
                if len(model._saved_files) == model._saved_files.maxlen:
                    # deque å·²æ»¡ï¼Œè¯´æ˜åˆšåˆšä¸¢å¼ƒäº†æœ€è€çš„
                    # æ‰‹åŠ¨åˆ é™¤è¿™ä¸ªæ–‡ä»¶
                    oldest = model._saved_files[0]
                    if os.path.exists(oldest):
                        os.remove(oldest)

                print(f"ğŸ’¾ ä¿å­˜æ›´ä¼˜è§£ {filename} (obj={obj})ï¼Œä¿ç•™æœ€è¿‘ {len(model._saved_files)} ä¸ªè§£")


    def solver(self):
        """
        å•è§£æ±‚è§£å‡½æ•°ï¼ˆå¯¹åº”åŸæ–‡ solver
        - æ„å»º Rebound-friendly æ¨¡å‹
        - æ±‚è§£æœ€ä¼˜è§£
        - è¿”å› solution, mds_solution, obj_value
        """
        self.__model = gp.Model()
        self.__state_vars = []

        self.construct_rebound_friend_model()
        model = self.__model


        # ===== åœ¨è¿™é‡ŒåŠ å…¨å±€çº¦æŸ & ç›®æ ‡ =====
        xor_match = gp.quicksum(self.__xor_match_vars)
        model.addConstr(xor_match <= self.__XorMatchLimit, name="xor_match_limit")##############è¿™é‡Œé™åˆ¶äº†1+1=0å‡ºç°çš„æ¬¡æ•°
        # model.setObjective(xor_match, gp.GRB.MINIMIZE)
        #########################æ³¨æ„è¿™é‡Œæ˜¯å…³äºxormatchçš„çº¦æŸå’Œç›®æ ‡å‡½æ•°

        model.Params.OutputFlag = 1 
        print("å˜é‡æ•°:", model.numVars)
        print("çº¦æŸæ•°:", model.numConstrs)
        print("ç›®æ ‡å‡½æ•°éé›¶æ•°:", model.numNZs)
        model.Params.MIPGap = 0.2      # gap å°äº 20% å°±åœ
        model.Params.TimeLimit = 60  


        # model.optimize()
        # model.optimize(self.stop_at_gap)
        model.optimize(self.my_callback)

        # åˆ é™¤ LP æ–‡ä»¶ï¼ˆå¯é€‰ï¼‰
        if os.path.exists(f"{self.__milp_lpfile_folder}/{self.__milp_lpfile_name}"):
            os.remove(f"{self.__milp_lpfile_folder}/{self.__milp_lpfile_name}")

        # æ‰“å°æ±‚è§£çŠ¶æ€
        print("Solver status:", model.status)

        # if model.status == gp.GRB.OPTIMAL:######################################
        # if model.status in [gp.GRB.OPTIMAL, gp.GRB.INTERRUPTED]:#############ä¸ºäº†æå‰ä¸­æ–­ä¹Ÿèƒ½è¾“å‡ºè§£
        if model.status in [gp.GRB.OPTIMAL, gp.GRB.TIME_LIMIT, gp.GRB.INTERRUPTED, gp.GRB.SUBOPTIMAL]:
            obj_value = model.getObjective().getValue()

            # åˆå§‹åŒ–è§£
            solution = [[[0 for _ in range(16)] for _ in range(self.__num_blocks)] for _ in range(self.__rounds)]
            mds_solution = [[[0 for _ in range(4)] for _ in range(self.__num_blocks)] for _ in range(self.__rounds)]

            for r in range(self.__rounds):
                for b in range(self.__num_blocks):
                    for j in range(16):
                        if self.__state_vars[r][b][j].xn == 1:
                            solution[r][b][j] = 1

            for r in range(len(self.__auxi_vars)):
                for b in range(self.__num_blocks):
                    for j in range(4):
                        if self.__auxi_vars[r][b][j].xn == 1:
                            mds_solution[r][b][j] = 1

            return solution, mds_solution, obj_value
        # else:
        #     return None
        elif model.status == gp.GRB.INFEASIBLE:
            print("æ¨¡å‹æ— è§£ (INFEASIBLE)ï¼Œå»ºè®®è¿è¡Œ model.computeIIS() æ£€æŸ¥å†²çªçº¦æŸã€‚")
            model.computeIIS()
            model.write("infeasible.ilp")

            return None
        else:
            print(f"æ¨¡å‹æœªæ‰¾åˆ°æœ€ä¼˜è§£,status={model.status}")
            return None


    def solvers(self):
        """
        å¤šè§£æ±‚è§£å‡½æ•°ï¼ˆå¯¹åº”åŸæ–‡ solvers
        - æ„å»º Rebound-friendly æ¨¡å‹
        - å¼€å¯è§£æ± æ¨¡å¼è·å–å¤šä¸ªå¯è¡Œè§£
        - è¿”å›æ‰€æœ‰è§£çš„ state_vars æ¿€æ´»æƒ…å†µ
        """
        self.__model = gp.Model()
        self.__state_vars = []
        # self._build_model()

        self.construct_rebound_friend_model()###########################
        model = self.__model
        model.setParam("PoolSearchMode", 2)
        ############ä¸‹é¢ä¸¤è¡Œæ˜¯åé¢è¡¥å……çš„
        model.setParam("PoolSolutions", 100)    # æœ€å¤šå­˜ 100 ä¸ªè§£
        model.setParam("PoolGap", 1.0)          # æ”¾å®½æœ€ä¼˜æ€§å®¹å¿åº¦ï¼ˆå¦åˆ™åªä¼šæ”¶é›†æœ€ä¼˜è§£ï¼‰å…è®¸å·®ä¸€ç‚¹çš„è§£ä¹Ÿèƒ½è¿›è§£æ± 

        model.optimize()

        num_solutions = model.getAttr("SolCount")
        print("Number of solutions:", num_solutions)

        result_solution = []

        for sol_idx in range(num_solutions):
            model.setParam("SolutionNumber", sol_idx)

            solution = [[[0 for _ in range(16)] for _ in range(self.__num_blocks)] for _ in range(self.__rounds)]
            for r in range(self.__rounds):
                for b in range(self.__num_blocks):
                    for j in range(16):
                        if self.__state_vars[r][b][j].xn == 1:
                            solution[r][b][j] = 1
            result_solution.append(solution)

        return result_solution

    
```

```python
import numpy as np

def assign_integers_to_solution(solution):
    """
    å°†solutionä¸­çš„é»‘æ ¼å­èµ‹å…·ä½“æ•´æ•°å€¼ï¼Œæ ¹æ®Permutationå’ŒMCæ‰©æ•£ã€‚
    ç™½æ ¼å­ä¿æŒ0ï¼Œé»‘æ ¼å­æŒ‰æ‰©æ•£é€»è¾‘èµ‹å€¼ã€‚
    """
    rounds = len(solution)
    num_blocks = len(solution[0])
    
    # ä¸ºæ¯ä¸ªé»‘æ ¼å­åˆ†é…æ•´æ•°æ ‡è®°
    next_value = 1
    int_solution = [[np.zeros((4, 4), dtype=int) for _ in range(num_blocks)] for _ in range(rounds)]
    
    # ä¿å­˜ä¸Šè½®çš„æ ¼å­æ•´æ•°æ˜ å°„ï¼Œä¾¿äºPropagation
    prev_blocks = [np.zeros((4, 4), dtype=int) for _ in range(num_blocks)]
    
    for r in range(rounds):
        for b in range(num_blocks):
            block = np.array(solution[r][b]).reshape(4, 4)
            new_block = np.zeros((4, 4), dtype=int)
            
            for i in range(4):
                for j in range(4):
                    if block[i, j] > 0.5:  # é»‘æ ¼å­
                        # å°è¯•ç»§æ‰¿ä¸Šä¸€è½®ç›¸åŒä½ç½®çš„å€¼
                        inherited = prev_blocks[b][i, j]
                        if inherited != 0:
                            new_block[i, j] = inherited
                        else:
                            # æ–°çš„æœªçŸ¥æ ¼å­ï¼Œåˆ†é…æ–°æ•´æ•°
                            new_block[i, j] = next_value
                            next_value += 1
                    # ç™½æ ¼å­ä¿æŒ0
            int_solution[r][b] = new_block
        
        # æ¨¡æ‹ŸPermutation/MCæ‰©æ•£ï¼Œå°†ä¸Šä¸€è½®æ•´æ•°æ˜ å°„ä¼ æ’­åˆ°ä¸‹ä¸€è½®
        # ç®€åŒ–ç‰ˆï¼šShiftRows + XORä¼ æ’­ï¼ˆå¯ä»¥æ ¹æ®å®é™…AES_TD.permutationè¿›ä¸€æ­¥ä¿®æ”¹ï¼‰
        prev_blocks = [np.copy(block) for block in int_solution[r]]
    
    return int_solution

# ä½¿ç”¨æ–¹æ³•
int_solution = assign_integers_to_solution(solution)

# å¯è§†åŒ–æ•´æ•°è½®è¡¨
fig, axes = plt.subplots(rounds, 12, figsize=(20, 10))
for r in range(rounds):
    for b in range(12):
        block = int_solution[r][b]
        grid = (block > 0).astype(int)
        axes[r, b].imshow(grid, cmap="gray_r", vmin=0, vmax=1)
        axes[r, b].set_title(f"R{r} B{b}")
        axes[r, b].set_xticks([])
        axes[r, b].set_yticks([])
        if b in highlight_blocks:
            for spine in axes[r, b].spines.values():
                spine.set_edgecolor('red')
plt.tight_layout()
plt.show()
```

```python
def solve_min_active_sboxes(aes_obj):
    """
    è¾“å…¥ï¼š
        aes_obj: MILPFindReboundPath ç±»çš„å®ä¾‹
    åŠŸèƒ½ï¼š
        - æ„å»ºæœ€å°æ´»è·ƒSç›’ MILP æ¨¡å‹
        - æ±‚è§£æœ€ä¼˜è§£
        - è¿”å›æ¯è½®æ¯å—çš„æ¿€æ´»æƒ…å†µ solution å’Œç›®æ ‡å€¼ obj_value
    """
    import gurobipy as gp
    
    # 1. æ¸…ç†æ¨¡å‹å’Œå˜é‡
    aes_obj._MILPFindReboundPath__model = gp.Model()
    aes_obj._MILPFindReboundPath__model.setParam(gp.GRB.Param.OutputFlag, False)
    aes_obj._MILPFindReboundPath__state_vars = []

    # 2. ç”Ÿæˆå˜é‡
    aes_obj.generate_variables()

    # 3. æ„å»ºæ‰€æœ‰è½®çš„ AES/XOR ç½®æ¢çº¦æŸ
    aes_obj.Permutation()

    # 4. è®¾ç½®ç›®æ ‡å‡½æ•°ï¼šæœ€å°åŒ–æ‰€æœ‰æ¯”ç‰¹æ´»è·ƒæ•°é‡
    aes_obj.set_object_function()

    # 5. åˆå§‹åŒ–çº¦æŸï¼šè‡³å°‘ä¸€ä¸ªæ¯”ç‰¹åœ¨ç¬¬0è½®è¢«æ¿€æ´»
    aes_obj.init_constraints()

    # 6. æ±‚è§£
    model = aes_obj._MILPFindReboundPath__model
    model.optimize()

    # 7. è§£æè§£
    if model.status == gp.GRB.OPTIMAL:
        rounds = aes_obj._MILPFindReboundPath__rounds
        num_blocks = aes_obj._MILPFindReboundPath__num_blocks
        block_size = aes_obj._MILPFindReboundPath__block_size
        solution = [[[0 for _ in range(block_size)] for _ in range(num_blocks)] for _ in range(rounds)]

        for r in range(rounds):
            for b in range(num_blocks):
                for i in range(block_size):
                    if aes_obj._MILPFindReboundPath__state_vars[r][b][i].xn == 1:
                        solution[r][b][i] = 1

        obj_value = model.getObjective().getValue()
        return solution, obj_value
    else:
        print("æœªæ‰¾åˆ°æœ€ä¼˜è§£")
        return None, None

rounds = 8
aes = MILPFindReboundPath(ciphername="AES_128", rounds=rounds, begin_round=0, end_round=0)
solution, obj = solve_min_active_sboxes(aes)

if solution:
    print("æœ€å°æ´»è·ƒSç›’æ•°é‡:", obj)
    # å¯è§†åŒ–
    import matplotlib.pyplot as plt
    import numpy as np

    fig, axes = plt.subplots(rounds, aes._MILPFindReboundPath__num_blocks, figsize=(24, 12))
    for r in range(rounds):
        for b in range(aes._MILPFindReboundPath__num_blocks):
            block = solution[r][b]
            grid = np.array(block).reshape(4, 4).T
            axes[r, b].imshow(grid, cmap="gray_r")
            axes[r, b].set_xticks([])
            axes[r, b].set_yticks([])
            axes[r, b].set_title(f"R{r} B{b}")
    plt.tight_layout()
    plt.show()

#######################æ­£ç¡®çš„åº”è¯¥æ˜¯105ä¸ª
```
